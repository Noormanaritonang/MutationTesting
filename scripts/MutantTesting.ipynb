```python
import glob, os
import pandas as pd
import csv
```

```python
listHasil = []
nameForTS = []
mutantScore = 0
allCountTS = 0

#Looping for Version
for version in range (1,10):
    FCI = 0
    NonFCI = 0
    #Looping for TestSuit
    for testSuit in range (1,6):
        pathOriginal = "D:\\Semester 7\\PUPL\\Week 11\\schedule\\outputs\\original\\TS%d\\"%(testSuit)
        pathVersion = "D:\\Semester 7\\PUPL\\Week 11\\schedule\\outputs\\version%d\\TS%d\\"%((version),(testSuit))
        
        TestCase = []
        nameForTC = []
        FCITS = 0
        NonFCITS = 0
        mutantScoreTS = 0
        allCountTC = 0
        # Looping untuk File yang ada didalam PathOriginal
        for File in os.listdir(pathOriginal):   # os.listdir metode mengembalikan folder tertentu berisi nama folder (file didalam folder TS Ori)
            if File.startswith('t'):            # File kita dimulai dengan char 't' 
                File = File.split("\\")         # menentukan pembatas untuk bagian string
                namaFile = File[len(File)-1]    # deskripsi dimana nama file akan dilooping  
                # Deskripsi fileOriginal
                fileOriginal = open(pathOriginal + namaFile, "r")  # fileOriginal itu diambil dari pathOriginal kemudian namaFile yang dibaca dengan fungsi r (read) 
                dataOriginal = fileOriginal.read()                 # Kemudian deskripsikan dataOri dari fileOri yg telah dikenalkan sebelumnya
                # Deskripsi fileVersion
                fileVersion = open(pathVersion + namaFile, "r")    # fileOriginal itu diambil dari pathOriginal kemudian namaFile yang dibaca dengan fungsi r (read) 
                dataVersion = fileVersion.read()                   # Kemudian deskripsikan dataVersion dari fileversion yg telah dikenalkan sebelumnya
    
                # Pengecekan untuk mencari FCI & Non FCI
                if(dataOriginal == dataVersion):   # Cek dataOri & dataVersion
                    listHasil.append("Non FCI")    # .append untuk Menambah nilai baru pada list (dgn nilai="NonFCI"), listHasil nama list
                    TestCase.append("Non FCI")     # menambah nilai baru pada list TestCase
                    NonFCI = NonFCI + 1            # Deskripsi NonFCI untuk masing2 test case
                    NonFCITS = NonFCITS + 1        # Deskripsi NonFCI untuk masing2 test Suit
                else: # Jika data Ori != dataVersion maka
                    listHasil.append("FCI")  # Kita tambahkan nilai baru pada listHasil
                    TestCase.append("FCI")   # Kita tambahkan nilai baru pada list TestCase
                    FCI = FCI + 1            # Deskripsi FCI untuk masing2 test case
                    FCITS = FCITS + 1        # Deskripsi FCI untuk masing2 test suit

                nameForTS.append(namaFile) # Kita tambahkan masing2 namaFile yang kita dapat ke list name
                nameForTC.append(namaFile) # Kita tambahkan masing2 namaFile yang kita dapat ke list nama
                fileOriginal.close()  # Close fileOriginal
                fileVersion.close()   # Close fileVersion
            # Menampilkan FCI & Non FCI pada masing2 Test Suit    
            print ("FCI for TS%d: %d and Non-FCI for TS%d: %d" %((testSuit),(FCITS),(testSuit),(NonFCITS)))
            # Total FCI & Non FCI pada testcase ori terhadap version di masing2 Test Suit 
            allCountTC = NonFCITS + FCITS
            # Mutant Score untuk masing2 testcase di semua Test Suit
            mutantScoreTS = 100*NonFCITS/(allCountTC - FCITS)
            # Menampilkan hasil perhitungan mutant score
            print("mutant score untuk TS%d: %d" %((testSuit),(mutantScoreTS)))
            # Fungsi pada python untuk melakukan export data ke csv
            PUPL = pd.DataFrame(nameForTC,TestCase)
            PUPL.to_csv("D:\\Semester 7\\PUPL\\Week 11\\schedule\\masing2TS\\Original Terhadap Version ke-%d di TS%d.csv" %((version),(testSuit)), index = True, header = False)
    
        #Menampillkan FCI & Non FCI pada testcase ori terhadap version di semua TestSuit
        print ("FCI : %d dan Non-FCI : %d" %((FCI),(NonFCI)))
        # Total FCI & Non FCI pada testcase ori terhadap version disemua Test Suit
        allCountTS = NonFCI + FCI
        # Mutant Score untuk testcase di semua Test Suit
        mutantScore = 100*NonFCI/(allCountTS - FCI)
        # Menampilkan hasil perhitungan mutant score
        print ("Mutant Score = %d" %(mutantScore))
        print(listHasil)
        # Fungsi pada python untuk melakukan export data ke csv
        df = pd.DataFrame(nameForTS,listHasil)
        df.to_csv("D:\\Semester 7\\PUPL\\Week 11\\schedule\\AllTS\\original Terhadap Version ke-%d di semua TS.csv" %(version), index = True, header = False)
```
